#### Чистим данные от знаков припинания, заглавных букв, табуляций и переходов на новую строку

```python
text = "".join([c for c in text if c.isalpha() or c.isdigit() or c ==' ']).lower()
#rstrip()
text
```

```python
key = int(input('Введите целочисленный ключ:\n '))
```

Создадим список самых используемых монограм в порядке убывания частоты использования

```python
alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
monograms = 'оеаинтсрвлкмдпуяыьгзбчйхжшюцщэфъё'
```

Сдвигаем буквы на   
__(index(symbol)  + pow(key,3)) % 33__  
Делим по модулю 33, т.к в алфавите всего 33 буквы

А пробелы просто добавляем

```python
encrypt_text = ''

for symbol in text:
    if symbol.isalpha(): # если символ входит в алфавит
        encrypt_text += alphabet[(alphabet.find(symbol) + pow(key,3)) % 33]
    else:
        encrypt_text += symbol
        
encrypt_text
```

## Частотный анализ монограм
#### Создадим словарь длинною в *alphabet* и заполним его нулями.  
После пробежимся по зашифрованному тексту и каждый раз, встречая одну и ту-же букву, увеличиваем значение ее частоты на 1 

```python
frequencies = {}

for symbol in alphabet: 
      frequencies[symbol] = 0
    
for symbol in encrypt_text:
    if symbol != ' ':
        frequencies[symbol] += 1
```

Сдвигаем список монограмм на значение *key*

```python
crypt_monograms = ''

for key, value in frequencies.items():
    i = 0
    while i < len(crypt_monograms) and value <= frequencies[crypt_monograms[i]]:
        i += 1
    
    crypt_monograms = crypt_monograms[:i] + key + crypt_monograms[i:]

print('Monograms:\t', monograms)
print('Crypt_Monograms:',crypt_monograms)
```

```python
decrypt_text1 = ''

for symbol in encrypt_text:
    if symbol.isalpha():
        decrypt_text1 += monograms[crypt_monograms.find(symbol)]
    else:
        decrypt_text1 += symbol
        
decrypt_text1
```

## Частотный анализ биграм

Пробегаемся по __тексту__ и составляем словарь биграм и их кол-во в тексте

```python
bigrams = {}

for i in range(len(text)-1):
    if text[i] == ' ' or text[i+1] == ' ':
        continue
        
    if text[i:i+2] not in bigrams.keys(): # если биграмма еше нет в словаре, то добавляем ее
        bigrams[text[i:i+2]] = 1
    else:
        bigrams[text[i:i+2]] += 1
        
print(bigrams)
```

Функия для склеивания частей текста, на которые мы разделили наш список монограм при помощи ключа

```python
def glue_parts(s, p1, p2):
    if p1 > p2:
        p1, p2 = p2, p1
        
    s1 = s[:p1]
    s2 = s[p1+1:p2]
    s3 = s[p2+1:]
    
    return s1 + s[p2] + s2 + s[p1] + s3
```

Пробегаемся по всем биграмам и сопоставляем зашифрованные биграмы незашифрованным.  
При помощи разбиения их на монограмы и их последующего анализа
